```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
print(len(purchase_data.SN.drop_duplicates()))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
n_items = len(purchase_data['Item ID'].drop_duplicates())
avg_price = purchase_data.Price.mean()
n_purchase = purchase_data.shape[0]
total_rev = purchase_data.Price.sum()

result_dict = {"Number of Items": [n_items], 
          "Average Price": [avg_price], 
          "Number of purchase":[n_purchase],
          "Total Revenue":[total_rev]}

result_df = pd.DataFrame(data = result_dict)
result_df['Average Price'] = result_df['Average Price'].map('${:,.2f}'.format)
result_df['Total Revenue'] = result_df['Total Revenue'].map('${:,.2f}'.format)
result_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender = purchase_data.loc[:,["SN","Gender"]].drop_duplicates()
n = len(gender)
result_df = gender.groupby('Gender').count()
result_df['Percentage of Players'] = result_df.iloc[:,0]/n * 100
result_df.rename(columns = {'SN':'Total Count'}, inplace = True)
result_df['Percentage of Players'] = result_df['Percentage of Players'].map('{:,.2f}%'.format)
result_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
import numpy as np
grp = purchase_data.groupby('Gender')
purchase_count = grp['Purchase ID'].count().values
avg_price = grp['Price'].mean().values
total =  grp['Price'].sum().values
avg_price_pp = total/grp['SN'].apply(lambda x: x.drop_duplicates().count())
result_df = pd.DataFrame(data = np.transpose(np.array([purchase_count, avg_price, total, avg_price_pp])), columns = ['Purchase Count',
                                                                                             'Average Purchase Price',
                                                                                             'Total Purchase Value',
                                                                                             'Avg Total Purchase per Person'])
result_df['Average Purchase Price'] = result_df['Average Purchase Price'].map('${:,.2f}'.format)
result_df['Total Purchase Value'] = result_df['Total Purchase Value'].map('${:,.2f}'.format)
result_df['Avg Total Purchase per Person'] = result_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)
result_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = pd.IntervalIndex.from_tuples([(0, 9), (9,14), (14,19), (19,24), (24,29), (29,34), (34,39), (39, float('inf'))])
purchase_data['age_category']= pd.cut(purchase_data.Age, bins, labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
idx = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

age = purchase_data.loc[:,["SN","age_category"]].drop_duplicates()
n = len(age)
result_df = age.groupby('age_category').count()
result_df['Percentage of Players'] = result_df.SN/n * 100
result_df.rename(columns = {'SN':'Total Count'}, inplace = True)
result_df['Percentage of Players'] = result_df['Percentage of Players'].map('{:,.2f}%'.format)
result_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grp = purchase_data.groupby('SN')
result_df = pd.DataFrame(grp['Price'].sum())
result_df.rename(columns = {'Price':'Total Purchase Value'}, inplace = True)
result_df['Average Purchase Price'] = grp['Price'].mean()
result_df['Purcahse Count'] = grp['Purchase ID'].count()

result_df.sort_values(by = 'Total Purchase Value', inplace = True, ascending = False)
result_df['Average Purchase Price'] = result_df['Average Purchase Price'].map('${:,.2f}'.format)
result_df['Total Purchase Value'] = result_df['Total Purchase Value'].map('${:,.2f}'.format)
result_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sub = purchase_data.loc[:, ['Item ID', 'Item Name','Price']]
grp = sub.groupby(['Item ID', 'Item Name'])
result_df = pd.DataFrame(sub.groupby(['Item ID', 'Item Name'])['Price'].apply(lambda x: x.iloc[0]))
result_df['Total Purchase Value'] = grp['Price'].sum()
result_df['Purchase Count'] = grp['Price'].count()
result_df.sort_values(by = 'Purchase Count', inplace = True, ascending = False)

result_df_r1 = result_df.copy()
result_df_r1['Total Purchase Value'] = result_df_r1['Total Purchase Value'].map('${:,.2f}'.format)
result_df_r1['Price'] = result_df_r1['Price'].map('${:,.2f}'.format)
result_df_r1.head(5)
```
